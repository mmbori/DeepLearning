```python
from PIL import Image

path = 'D:/Image/LR/3/32/ESP-장X지-00000000000-EOS200DII-껍데기 있는 계란-20210203_IMG_3328_x2.jpg'

image = Image.open(path)

width, height = image.size

print("width : %d, hieght : %d" % (width,height))

image.show()
```

```python
img_resize = image.resize((int(width*0.5), int(height*0.5)))
```

```python
# BICUBIC 외에도 다른 기법 전부 적용해봄
```

```python
resize_bicubic = img_resize.resize((width, height), Image.BICUBIC)
#resize_bicubic.show()
resize_bicubic.save('저장경로', 'jpg')
```

```python
resize_nearest = img_resize.resize((width, height), Image.NEAREST)
#resize_nearest.show()
resize_nearest.save('저장경로', 'jpg')
```

```python
resize_box = img_resize.resize((width, height), Image.BOX)
#resize_box.show()
resize_box.save('저장경로', 'jpg')
```

```python
resize_hamming = img_resize.resize((width, height), Image.HAMMING)
#resize_hamming.show()
resize_hamming.save('저장경로', 'jpg')
```

```python
resize_lanczos = img_resize.resize((width, height), Image.LANCZOS)
#resize_lanczos.show()
resize_lanczos.save('저장경로', 'jpg')
```
